```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
print("*******1st question********")
print()
img = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)

def get_median(arr):
    import statistics
    temp = []
    for i in arr:
        for j in i:
            temp.append(j)
            
    return statistics.median(temp)

def median_filter(arr):
    test = img.copy()
    padded = cv.copyMakeBorder(img, 2, 2, 2, 2, cv.BORDER_REPLICATE, None, value = 0)
    
    for i in range(2,img.shape[0]+2):
        for j in range(2,img.shape[1]+2):
            part = padded[i-2:i+3,j-2:j+3]
            new_value = get_median(part)
            test[i-2,j-2] = new_value

    return test

filtered_image_opencv = cv.medianBlur(img,5)

filtered_image = median_filter(img)

print("abs(my filter-openCV's filter).sum() :",abs(filtered_image-filtered_image_opencv).sum())
print()
print("*******2nd question********")
print()
#2. soru

lena_grayscale = cv.imread("lena_grayscale_hq.jpg", cv.IMREAD_GRAYSCALE)

noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)
box_filtered = None
box_filtered = cv.boxFilter(noisyImage, 0, (5,5), box_filtered, (-1,-1), True, cv.BORDER_CONSTANT)

noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)
gaussian = cv.GaussianBlur(noisyImage,(7,7),0)

noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)
median = cv.medianBlur(noisyImage,5)


print("OpenCV's box filter output PSNR:",cv.PSNR(lena_grayscale, box_filtered))
print("OpenCV's Gaussian filter output PSNR:",cv.PSNR(lena_grayscale, gaussian))
print("OpenCV's median filter output PSNR:",cv.PSNR(lena_grayscale, median))
print()
print("*******3rd question********")
print()
#3. soru

def get_median_weighted(arr):
    import statistics
    temp = []
    temp.append(arr[2,2])
    temp.append(arr[2,2])
    for i in arr:
        for j in i:
            temp.append(j)
    return statistics.median(temp)

def weighted_median_filter(img):
    test = img.copy()
    padded = cv.copyMakeBorder(img, 2, 2, 2, 2, cv.BORDER_REPLICATE, None, value = 0)
    
    for i in range(2,img.shape[0]+2):
        for j in range(2,img.shape[1]+2):
            part = padded[i-2:i+3,j-2:j+3]
            new_value = get_median_weighted(part)
            test[i-2,j-2] = new_value

    return test

lena_grayscale = cv.imread("lena_grayscale_hq.jpg", cv.IMREAD_GRAYSCALE)

noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)
box_filtered = None
box_filtered = cv.boxFilter(noisyImage, 0, (5,5), box_filtered, (-1,-1), True, cv.BORDER_CONSTANT)

noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)

gaussian = cv.GaussianBlur(noisyImage,(7,7),0)
noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)

median = cv.medianBlur(noisyImage,5)
noisyImage = cv.imread("noisyImage.jpg", cv.IMREAD_GRAYSCALE)

weighted_filtered = weighted_median_filter(noisyImage)

median_filtered_psnr = cv.PSNR(lena_grayscale, filtered_image)
print("Own median filter output PSNR: ",median_filtered_psnr)
box_psnr = cv.PSNR(lena_grayscale, box_filtered)
print("OpenCV's box filter output PSNR:",box_psnr)
gaussian_psnr = cv.PSNR(lena_grayscale, gaussian)
print("OpenCV's Gaussian filter output PSNR:",gaussian_psnr)
median_psnr = cv.PSNR(lena_grayscale, median)
print("OpenCV's median filter output PSNR:",median_psnr)
weighted_psnr = cv.PSNR(lena_grayscale, weighted_filtered)
print("Own center weighted median filter output PSNR:",weighted_psnr)

#4. soru
print()
print("*******4th question********")
print()
best_image = weighted_filtered.copy()
cv.imshow("my new x result",best_image)
cv.waitKey(0)
cv.destroyAllWindows()
def my_filter(img):
    import random
    m,n = img.shape
    
    for x in range(m):
        for y in range(n):
            if(img[x,y] < 240 and img[x,y] > 200):
                img[x,y] = img[x,y]+5
            elif(img[x,y] <= 100 and img[x,y] > 15):
                img[x,y] = img[x,y]-5
            elif(img[x,y] <= 200 and img[x,y] > 150):
                img[x,y] = img[x,y]-5
            else:
                img[x,y] = img[x,y]+5
                
    return cv.medianBlur(img,3)
    
filtered = my_filter(best_image)
lena_grayscale = cv.imread("lena_grayscale_hq.jpg", cv.IMREAD_GRAYSCALE)

print("my new filter's PSNR:",cv.PSNR(lena_grayscale, filtered))


cv.imshow("my new filter result",filtered)
cv.waitKey(0)
cv.destroyAllWindows()

```
